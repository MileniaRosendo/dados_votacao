```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/csec_2t_AM_291020221210.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=';')

# Visualizar as primeiras linhas para inspeção inicial
print(data.head())
```

```python
print(data.info())
print(data.isnull().sum())
```

```python
data['NM_MUNICIPIO'] = data['NM_MUNICIPIO'].str.upper().str.strip()
data['NR_ZONA'] = data['NR_ZONA'].astype(str).str.strip()
```

```python
data['DT_GERACAO'] = pd.to_datetime(data['DT_GERACAO'], format='%d/%m/%Y')
data['DT_CARGA_URNA_ESPERADA'] = pd.to_datetime(data['DT_CARGA_URNA_ESPERADA'], format='%d/%m/%Y %H:%M:%S')
data['HH_GERACAO'] = pd.to_datetime(data['HH_GERACAO'], format='%H:%M:%S').dt.time
```

```python
data['NR_URNA_ESPERADA'] = pd.to_numeric(data['NR_URNA_ESPERADA'], errors='coerce')
q1 = data['NR_URNA_ESPERADA'].quantile(0.25)
q3 = data['NR_URNA_ESPERADA'].quantile(0.75)
iqr = q3 - q1
filtro = (data['NR_URNA_ESPERADA'] >= (q1 - 1.5 * iqr)) & (data['NR_URNA_ESPERADA'] <= (q3 + 1.5 * iqr))
data = data[filtro]
print(data)
```

```python
data.to_csv('/content/dados_tratados.csv', index=False, encoding='ISO-8859-1', sep=';')
```
